## Numerical evaluation of the derivative of a function from definition

Import python libraries

```python
import numpy as np  #llibreries que importem sempre
import matplotlib.pyplot as plt
```

definir una funció

```python
def my_function(u):
    w=2.0*np.pi
    f=np.sin(w*u)
    return(f)
```

```python
x_list = np.arange(0.0, 2.0, 0.001)
num_punts=len(x_list)
print(num_punts)
```

Evaluate the function (example: sin(w x))

```python
y=my_function(x_list)
print(y)
```

Create empty arrays for the derivative (x and y)

```python
x2=[]
y2=[]
```

Loop over these points

```python
i=0     #while vol dir que repeteixi el calcul mentre es donin les condicions que li he donat, és un contador
while i<num_punts-1:
     print(i)
     derivative=(y[i+1]-y[i])/(x_list[i+1]-x_list[i])
     y2.append(derivative)      #append serveix per afegir valors en una llista, la qual hem creat previament i és buida
     x2.append(x_list[i])
     i=i+1
#aquest codi d'aqui és el que calcula la derivada com a tal, la resta era preparació i representació. Aquest és el codi de la derivada
```

Plot the results

```python
plt.plot(x_list, y, 'r', x2, y2, 'b')
plt.xlabel('x')
plt.ylabel('f and df/dx')
plt.title('Numerical calculation of derivative')
plt.legend(['f(x)', 'df/dx'])
plt.show()
```
