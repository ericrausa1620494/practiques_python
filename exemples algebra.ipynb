podem treballar amb matrius tambe

```python
M=np.array([[2,2],[1,3]])
print(M)
```

```python
e=M[0,1]
print(e)
```

```python
v=M[1,:] #els dos punts volen dir que agafi tots els numeros, és un comodin, en aquest cas li diem que el vector és tota la 2a fila
print(v)
```

```python
print (M) 
```

```python
#per trobar valor si vectors propis, li demanem linalg.eig(Matriu)
valors,vectors=np.linalg.eig(M)
print("valors propis", valors)
print("vectors propis", vectors)
```

```python
u=vectors[:,0]
print(u)
```
